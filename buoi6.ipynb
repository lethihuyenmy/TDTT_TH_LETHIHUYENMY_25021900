```python
def max(a,b):
    if a>b:
        return a
    return b
a,b=map(int,input().split())
print(max(a,b))
```

```python
def swap(a,b):
    a,b=b,a
    return f"{a} {b} "
a,b=map(int,input().split())
print(swap(a,b))
```

```python
def check_snt(a):
  if(a==0 or a==1): return False
  for i in range(2,int(a**(1/2))+1,1):
    if(a%i ==0): return False
  return True
n=-1
while (n<=0):
  n =int(input())
if(check_snt(n)): print ("True")
else: print("False")
```

```python
def perfect(n):
    sum=0
    for i in range(1,n):
        if n%i==0:
            sum+=i
    if sum==n:
        return True
    return False        
n=int(input())
print(perfect(n))
```

```python
def index(n,k):
    c=-1
    for i in range(len(n)):
        if n[i]==k:
            c=i
            break
        else:
            c=-1
    return c    
n=list(map(int,input().split()))
k=int(input())
print(index(n,k))
```

```python
n=int(input("Nhập số cần tính: "))
def giaithua(n):
    if n == 0:
        return 1
    return n * giaithua(n-1)
print(giaithua(n))
```

```python
def cal(a,b,c):
    if c=="-":
        return f"{a-b:.2f}"
    if c=="+":
        return f"{a+b:.2f}"
    if c=="*":
        return f"{a*b:.2f}"
    if c=="/":
        return f"{a/b:.2f}"
a, b, c=input().split() 
a,b=float(a),float(b)   
print(cal(a,b,c))
```

```python
def hamming_distance(s1, s2):
    return (s1 ^ s2).bit_count()
s1=int(input())
s2=int(input())
print(hamming_distance(s1,s2))
```

```python
def sum(a):
    sum=0
    for i in a:
        sum+=int(i)
    return sum
n=input() 
print(sum(n))   
```
