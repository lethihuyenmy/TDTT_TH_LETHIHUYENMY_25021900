```python
#W6A4
string=list(input().split())
m={}
for s in string:
    lst=[]
    str1=""
    str2=""
    check=False
    for u in s:
        if(u==":"):
            check=True
            continue
        if(not check):
            str1+=u
        else: str2+=u
    if str1 not in m:
        lst.append(str2)
    else:
        lst=m[str1]
        lst.append(str2)
    m[str1]=lst
print(m)

                
```

```python
#W6A10
n=list(map(int,input().split()))
k=int(input())
m=[]
for i in range(len(n)):
    for j in range(i+1,len(n)):
        list=[]
        if n[i]+n[j]==k:
            list.append(n[i])
            list.append(n[j])
        if list not in m and len(list)!=0:
            a=tuple(list)
            m.append(a)        
print(m)
```

```python
#W6A17
n=int(input())
a=[]
for i in range(n):
  row=[]
  x=map(int,input().split())
  for t in x:
    row.append(t)
  a.append(row)
print("Đường chéo chính")
for i in range(n):
  print(a[i][i],end=" ")
print()
print("Đường chéo phụ")
for i in range(n):
  print(a[i][n-1-i],end=" ")
```

```python
#W6A18
n,m=map(int,input().split())

a=[]
for i in range(n):
  row=[]
  x=map(int,input().split())
  for t in x:
    row.append(t)
  a.append(row)
k=int(input())
sum_row=0
sum_col=0
for i in range(n):
  sum_col+=a[i][k-1]
for j in range(m):
  sum_row+=a[k-1][j]
print(sum_row,sum_col)
```
