```python
def reverse(path):
    file = open(path, "r")
    n = file.readline().strip()
    file.close()
    reverse_str = n[::-1]
    return int(reverse_str)
path = "test.txt"
print(reverse(path))
```

```python
def canWinNim(path):
    file = open(path, "r")
    n = file.readline().strip()
    file.close()
    n=int(n)
    if (n == 4):
        return False
    return True
path = "test.txt"
print(canWinNim(path))
```

```python
#c1
def housesofhogwarts(path):
    file = open(path, "r")
    n = file.readlines().strip()
    file.close()
    for i in range(1, len(n)):
        print(n[i])
    file.close()    
path = "test.txt"
housesofhogwarts(path)        
```

```python
#c2
def housesofhogwwarts(path):
    file = open(path, "r")
    for line in file:   # không loại bỏ hàng đầu đọc hết các dòng
        print(line)
    file.close()    
path = "test.txt"
housesofhogwarts(path)
```

```python
def count(path, a):
    file = open(path, "r")
    n = file.readline().strip().lower()
    file.close()
    count = 0
    for i in n:
        if i==a:
            count+=1
    return(count)
path = "test.txt"
a=input()
print(count(path, a))        
```
