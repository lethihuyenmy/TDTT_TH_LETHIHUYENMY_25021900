```python
#W11A11
import re
def check_re(n):
    try:
        n=int(input())
        for i in range(n):
            a=input()
        try:
            re.compile(a)
            print(True)
        except re.error:
            print(False)
    except:
        pass            
if __name__ == "__main__" :
    check_re(n)

```
