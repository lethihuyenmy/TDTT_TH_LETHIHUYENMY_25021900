```python
#W9A7
n=list(map(int, input("Nhập vị trí cây: ").split()))
k=int(input(" Nhập số cấy để điền vào: "))
cocay=[]
h=0
for i in range(len(n)):
    if n[i]==1:
        cocay.append(i)
if len(cocay)>0:
    q=len(n)-1-cocay[len(cocay)-1]       
if len(n)>0 and len(cocay)==0:
    if len(n)%2==0:
        h+=len(n)/2
    else:
        h+=len(n)/2+1
if len(cocay)>0:
    if cocay[0]>=2 :
        h+=cocay[1]//2
    if q>=2:  
        h+=q//2    
    for i in range(len(cocay)):
        if i+1<=len(cocay)-1:
            a=cocay[i+1]-cocay[i]
            if a>=3 :
                if a%2==0:
                    h+= a//2-1
                else:
                    h+= a//2
if h>=k:
    print("True")
else:
    print("False")                                              
```

```python
#W9A8
class destiny:
    def __init__(self, a, b):
        self.a=a
        self.b=b
    def R(self):
        self.a+=1
        c=[self.a, self.b]
        return c
    def L(self):
        self.a-=1
        c=[self.a, self.b]
        return c
    def U(self):
        self.b+=1
        c=[self.a, self.b]
        return c
    def D(self):
        self.b-=1
        c=[self.a, self.b]
        return c
a=0
b=0
destiny=destiny(a,b)
n=input("Nhập bước đi: ") 
m=[]
for i in n:
    if i=="R":
        c=destiny.R()
        c=list(c)
        a=c[0] 
        b=c[1] 
        m.append(c)
    if i=="L":
        c=destiny.L()
        c=list(c)
        a=c[0] 
        b=c[1] 
        m.append(c)  
    if i=="U":
        c=destiny.U()
        c=list(c)
        a=c[0] 
        b=c[1] 
        m.append(c)      
    if i=="D":
        c=destiny.D()
        c=list(c)
        a=c[0] 
        b=c[1] 
        m.append(c)    
for i in range(len(m)-1):
    if m[i][0]==m[len(m)-1][0] and m[i][1]==m[len(m)-1][1]:
        print("True")
        break
    else:
        print("False")
        break
```

```python
#W9A9
def Median(nums):
    sum=0
    for i in nums:
        sum+=i
    return sum//5    
nums=list(map(int,input().split()))
if len(nums)!=5:
    nums=list(map(int,input().split()))
print(Median(nums))
```
