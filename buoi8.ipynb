```python
#W8A1
class Cylinder:
    def __init__(self, radius, h):
        self.radius = radius
        self.h = h
    def dien_tich_toan_phan(self):
        return 2 * 3.14 * self.radius * (self.radius + self.h)
    def the_tich(self):
        return 3.14 * self.radius * self.radius * self.h
radius, h =list(map(float,input().split()))
cylinder = Cylinder(radius, h)
print(f"{cylinder.dien_tich_toan_phan():.2f}")
print(f"{cylinder.the_tich():.2f}")    
```

```python
#W8A2
class Date:
    def __init__(self, dd, mm, yyyy):
        self.dd = dd
        self.mm = mm
        self.yyyy = yyyy        
    def date(self):
        c=0
        if yyyy%400==0 or (yyyy%4==0 and yyyy%100!=0):
            c=1 
        a=[1, 3, 5, 7, 8, 10, 12]
        b=[4, 6, 9, 11]
        if (mm in a and 1<=dd<=30) or (mm in b and 1<=dd<=29):
            return f"{dd+1}/{mm}/{yyyy}"
        if (mm in b and dd==30) or (mm in a and dd==31 and mm!=12):
            return f"1/{mm+1}/{yyyy}"
        if mm==12 and dd==31:
            return f"1/1/{yyyy+1}"
        if (c==1 and dd<=28 and mm==2) or (c==0 and dd<=27 and mm==2):
            return f"{dd+1}/{mm}/{yyyy}"
        if (c==1 and dd==29 and mm==2) or (c==0 and dd==28 and mm==2):
            return f"1/{mm+1}/{yyyy}"
        else:
            return "INVALID"
dd, mm, yyyy= map(int, input().split("/"))
dates = Date(dd, mm, yyyy) 
print(dates.date())

```

```python
#W8A3
from math import sqrt
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def point(self):
        if self.x==0==self.y:
            return f"M là tọa độ gốc, khoảng cách là 0"
        if self.x!=0 and self.y==0:
            return f"M nằm trên trục hoành, khoảng cách tới O là {self.x:.2f}"
        if self.x==0 and self.y!=0:
            return f"M nằm trên trục tung, khoảng cách tới O là {self.y:.2f}"
        if self.x!=0 and self.y!=0:
            a = sqrt(self.x**2 +self.y**2)
            return f"M không là gốc tọa độ, không nằm trên trục tung, trục hoành, khoảng cách tới O là {a:.2f}"
x, y = map(float, input().split())
points=Point(x, y) 
print(points.point())       
```

```python
#W8A7
class Fraction:
    def __init__(self, num, den):
        self.num =num
        self.den = den
    def rutgon(self):    
        if self.num<=self.den:
            max=0
            for i in range (2,self.num+1):
                if self.num % i==0 and self.den%i==0:
                    if max<i:
                        max=i
            self.num/=max
            self.den/=max
        else:
            for i in range (2,self.den+1):
                if self.num % i==0 and self.den%i==0:
                    self.num/=i
                    self.den/=i            
        return f"{self.num:.0f}/{self.den:.0f}"            
    def add(self, other):
        self.num = other.num * self.den + other.den * self.num
        self.den = other.den * self.den              
        return Fraction.rutgon(self)
    def sub(self, other):
        self.num = - other.num * self.den + other.den * self.num
        self.den = other.den * self.den              
        return Fraction.rutgon(self)
    def div(self, other):
        self.num = other.den * self.num 
        self.den = other.num * self.den              
        return Fraction.rutgon(self)
    def mul(self, other):
        self.num = other.num * self.num 
        self.den = other.den * self.den              
        return Fraction.rutgon(self)
a, b, op, c, d= input().split()
a=int(a)
b=int(b)
c=int(c)
d=int(d)
fraction1 = Fraction(a, b) 
fraction2 = Fraction(c, d)
if op == "+":
    print (fraction1.add(fraction2))    
if op == "-":
    print(fraction1.sub(fraction2))
if op == "/":
    print(fraction1.div(fraction2))
if op == "*":
    print(fraction1.mul(fraction2))            
```

```python
#W8A8
class BankAccount:
    def __init__(self,owner, balance):
        self.owner = balance
        self.balance = owner
    def deposit(self, k):
        self.owner= self.owner + k
        return self.owner
    def withdraw(self, k):
        self.owner=self.owner -k
        return self.owner
owner, balance= input().split()
balance=int(balance)
bankaccount = BankAccount(owner, balance)  
a= int(input("Nhập số lần: ")) 
for i in range(a):
    b= input("Yêu cầu: ")
    if 'DEPOSIT' in b:
        b=b.strip("DEPOSIT ")     
        b=int(b)
        owner= bankaccount.deposit(b)
    else:
        b=b.strip("WITHDRAW ")     
        b=int(b)
        owner = bankaccount.withdraw(b)    
        if owner<0:
            owner+=b
print(owner)
```

```python
#W8A9 
class student:
    def __init__(self,name):
        self.name=name
        self.scores=[]
    def score(self,score):
        self.scores.append(score)
    def avg(self): 
        if len(self.scores)==0:
            return 0   
        return sum(self.scores)/len(self.scores)
    def rank(self):
        self.score=self.avg()
        if self.score>=8:
            return "Excellent"
        if 6.5<=self.score <8:
            return "Good"   
        if 5<=self.score<6.5:
            return "Average"
        if 0<=self.score<5:
            return "Poor"
name=input()
n=int(input())
student=student(name)
for i in range(n):
    sub,score=input().split()
    score=float(score)
    student.score(score)

print(f"{name} {student.avg():.2f} {student.rank()} ")        
```

```python
#W8A11
from math import *
def vector(a,b):
    return [b[0]-a[0],b[1]-a[1],b[2]-a[2]]
def cross(u,v):
    return [u[1]*v[2]-u[2]*v[1],u[2]*v[0]-u[0]*v[2],u[0]*v[1]-u[1]*v[0]]
def dot(u,v):
    return (u[0]*v[0]+u[1]*v[1]+u[2]*v[2])
def no(u):
    return sqrt(dot(u,u))   
A=list(map(float, input().split()))
B=list(map(float, input().split()))
C=list(map(float, input().split()))
D=list(map(float, input().split()))
AB=vector(A,B)
AC=vector(A,C)
BC=vector(B,C)
BD=vector(B,D)
n1=cross(AB,AC)
n2=cross(BC,BD)
cos=abs(dot(n1,n2)/(no(n1)*no(n2)))
phi=degrees(acos(cos))
print(f"{phi:.2f}")

```
